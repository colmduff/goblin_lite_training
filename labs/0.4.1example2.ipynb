## Multiple Instances of GOBLIN lite

```python
from goblin_lite.goblin import ScenarioRunner
import os
import pandas as pd
```

### Setting Paths 
When running GOBLIN as a single instance, the standard approach goblin will take is to clear the internal database of the previous results and proceed to generate new ones based on the users provided configuration file. 

However, sometimes it may be necessary to run a vast number of scenarios in parallel. To do this, we can run mutliple instances of goblin at the same time. It is just a class at the end of the day. 

But, how do we get around the fact that each instance will wipe the internal database every time? 

The approach here is to specify a database directory for each instance. 

Below, we generate a path for each scenario input, where input folders for CBM will be saved, as well as the database associated with each instance. 

```python
# Database base path
db_path = "../data/DATABASE_BATCH/"
# Create the main database directory if it does not exist
if not os.path.exists(db_path):
    os.makedirs(db_path)

for i in range(5):
    # Construct the path for the new subdirectory
    new_dir_path = os.path.join(db_path, str(i))
    # Create the subdirectory if it does not exist
    if not os.path.exists(new_dir_path):
        os.makedirs(new_dir_path)
```

### Set paths and variables as usual

```python
input_path ="../data/input"
output_path = "../data/output"

ef_country = "ireland" #keep lower case, I have not implemented checks for this yet

baseline_year = 2020
target_year = 2050

forest_config = os.path.join(input_path, "forest_config", "cbm_factory.yaml")
```

### Create instances 

Now we create a dictionary of the instances

```python
runner_instances = {}

for i in range(5):  # This will create instances from 0 to 2

    runner_instance = ScenarioRunner(
        ef_country, baseline_year, target_year,
        f"{input_path}/scenario_set_{i}.json", forest_config,
        f"{os.path.join(db_path, str(i))}/instance_{i}.db"
    )
    runner_instances[f'runner_{i}'] = runner_instance
```

Then loop through and run each instance.

```python
for runner in runner_instances.values():
    runner.run_scenarios()
```
